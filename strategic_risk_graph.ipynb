```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the Excel file
data = pd.read_excel('/content/drive/MyDrive/climate/strategic_risk3.xlsx')

# Check if both columns 'ROE' and 'ROA' exist in the dataset
if 'ROE' in data.columns and 'ROA' in data.columns:
    # Plot the ROE vs ROA graph
    plt.figure(figsize=(10, 6))
    plt.scatter(data['ROA'], data['ROE'], color='blue', alpha=0.7, edgecolor='k', label='ROE vs ROA')
    plt.title('ROE vs ROA', fontsize=16)
    plt.xlabel('ROA (Return on Assets)', fontsize=14)
    plt.ylabel('ROE (Return on Equity)', fontsize=14)
    plt.grid(alpha=0.3)
    plt.legend()
    plt.tight_layout()
    plt.show()
else:
    print("The dataset does not contain the columns 'ROE' and 'ROA'.")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
data = pd.read_excel('/content/drive/MyDrive/climate/strategic_risk3.xlsx')

# Check if both columns 'ROE' and 'ROA' exist in the dataset
if 'ROE' in data.columns and 'ROA' in data.columns:
    # Set the style of the seaborn plot
    sns.set_theme(style="whitegrid")

    # Create a figure and axes
    plt.figure(figsize=(10, 6))

    # Scatter plot with regression line
    sns.regplot(x='ROA', y='ROE', data=data, scatter_kws={'color': 'blue', 'alpha': 0.6}, line_kws={'color': 'red', 'label': 'Regression Line'})

    # Customize the plot
    plt.title('Relationship Between ROE and ROA', fontsize=16)
    plt.xlabel('ROA (Return on Assets)', fontsize=14)
    plt.ylabel('ROE (Return on Equity)', fontsize=14)
    plt.legend()
    plt.tight_layout()

    # Show the plot
    plt.show()
else:
    print("The dataset does not contain the columns 'ROE' and 'ROA'.")
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_excel('/content/drive/MyDrive/climate/strategic_risk3.xlsx')

# Check if both columns 'ROE' and 'Total liability' exist in the dataset
if 'ROE' in data.columns and 'Total liability' in data.columns:
    # Set the style of the seaborn plot
    sns.set_theme(style="ticks")

    # Pairplot for 'ROE' and 'Total liability'
    sns.pairplot(data[['ROE', 'Total liability']], kind='scatter')

    # Customize the plot
    plt.suptitle('Pairplot: ROE vs Total Liability', fontsize=16)
    plt.tight_layout()
    plt.show()
else:
    print("The dataset does not contain the columns 'ROE' and 'Total liability'.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_excel('/content/drive/MyDrive/climate/strategic_risk3.xlsx')

# Check if both columns 'ROE' and 'Strategic Risk Taking' exist in the dataset
if 'ROE' in data.columns and 'Strategic Risk Taking' in data.columns:
    # Create a scatter plot
    plt.figure(figsize=(10, 6))
    scatter = plt.scatter(data['Strategic Risk Taking'], data['ROE'], c=data['Strategic Risk Taking'], cmap='viridis', edgecolors='w', alpha=0.7)

    # Add color bar
    plt.colorbar(scatter, label='Strategic Risk Taking')

    # Customize the plot
    plt.title('ROE vs Strategic Risk Taking', fontsize=16)
    plt.xlabel('Strategic Risk Taking', fontsize=14)
    plt.ylabel('ROE (Return on Equity)', fontsize=14)

    # Show the plot
    plt.tight_layout()
    plt.show()
else:
    print("The dataset does not contain the columns 'ROE' and 'Strategic Risk Taking'.")
```
